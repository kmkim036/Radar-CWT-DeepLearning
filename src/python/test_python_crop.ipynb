```python
def preprocessing(data): 
    image = np.zeros(shape=(len(data), 81, 1920)) # 빈 넘파이형태 (20*400)
    for i, row in enumerate(data.index):
        df = np.fromstring(data['pixels'][row], dtype=int, sep=' ') # 공백으로 구분된 데이터를 넘파이로 변경
        df = np.reshape(df, (81, 1920)) 
        image[i] = df
    return image
```

```python
x_train = preprocessing(data)
```
